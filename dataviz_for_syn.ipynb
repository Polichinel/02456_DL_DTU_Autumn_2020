```python
import numpy as np
import pickle
import matplotlib.pyplot as plt

import pandas as pd
import geopandas as gpd
```

```python
pkl_file = open('ucpd_grid.pkl', 'rb')
ucpd_grid = pickle.load(pkl_file)
pkl_file.close()
```

```python
pkl_file = open('ucpd_vol.pkl', 'rb')
ucpd_vol = pickle.load(pkl_file)
pkl_file.close()
```

```python
def plotPattern(ucpd_grid, fig_title):

    x = ucpd_grid[ucpd_grid['log_best'] > 0]['xcoord'] # -180 - 180
    z = ucpd_grid[ucpd_grid['log_best'] > 0]['ycoord'] # -90 - 90
    y = ucpd_grid[ucpd_grid['log_best'] > 0]['year']
    c = ucpd_grid[ucpd_grid['log_best'] > 0]['log_best']


    fig = plt.figure(figsize = [30,15])
    # fig.tight_layout()
    plt.subplots_adjust(hspace = -0.001, wspace = -0.001)


    #  First subplot-------------------------------------
    ax = fig.add_subplot(2, 2, 1, projection='3d')

    ax.scatter(x,y,z, c=c, alpha = 0.1)
    ax.set_xlabel('longitude')
    ax.set_ylabel('Years')
    ax.set_zlabel('latitude')

    ax.set_yticks([])

    ax.set_zlim(-90, 90)
    ax.set_xlim(-180, 180)

    ax.view_init(elev=0, azim=-90)
    ax.dist=8
    plt.title('Plot A')


    # second subplot-------------------------------------
    ax = fig.add_subplot(2, 2, 2, projection='3d')

    ax.scatter(x,y,z, c=c, alpha = 0.1)
    ax.set_xlabel('longitude')
    ax.set_ylabel('Years')
    ax.set_zlabel('latitude')

    ax.set_xticks([])

    ax.set_zlim(-90, 90)
    ax.set_xlim(-180, 180)

    ax.view_init(elev=0, azim=-0)
    ax.dist=8
    plt.title('Plot B')


    # third subplot -------------------------------------

    ax = fig.add_subplot(2, 2, 3, projection='3d')

    ax.scatter(x,y,z, c=c, alpha = 0.1)
    ax.set_xlabel('longitude')
    ax.set_ylabel('Years')
    ax.set_zlabel('latitude')

    ax.set_zticks([])

    ax.set_zlim(-90, 90)
    ax.set_xlim(-180, 180)

    # ax.view_init(elev=90, azim=-90)
    ax.view_init(elev=-90, azim=-90)

    ax.dist=8
    plt.title('Plot C')


    # fourth subplot -------------------------------------

    ax = fig.add_subplot(2, 2, 4, projection='3d')

    ax.scatter(x,y,z, c=c, alpha = 0.1)
    ax.set_xlabel('longitude')
    ax.set_ylabel('Years')
    ax.set_zlabel('latitude')

    ax.set_zlim(-90, 90)
    ax.set_xlim(-180, 180)

    ax.view_init(elev=45, azim=-45)
    ax.dist=10
    plt.title('Plot D')

    plt.savefig(fig_title, bbox_inches="tight")

    plt.show()
```

```python
plotPattern(ucpd_grid, "conflict_patterns.jpeg")
```
